```python
from bs4 import BeautifulSoup as bs 
import requests 
```

```python
r = requests.get("https://www.jumia.com.ng/fashion-new-mens-sneakers-thick-soled-casual-shoes-white-47608073.html")
#convert to a beautiful soup object 
soup = bs(r.content) 

#print out the HTML 
contents = soup.prettify() 
print(contents) 
```

```python
product_reviews = soup.find(class="Product Reviews") 
print(product_reviews.prettify())
```
